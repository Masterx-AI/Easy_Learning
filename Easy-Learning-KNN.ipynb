```python
#Importing necessary libraries

import copy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Video, clear_output
```

```python
#Scatter Points
a=np.random.randint(50, size=20)
b=np.random.randint(50, size=20)

#Random Start points for clusters
c=np.random.randint(50, size=[2,2])

#distance
d=np.zeros(20)
d1=np.zeros(20)
d2=np.zeros(20)
#d1=np.random.randint(1, size=[20,2])
#d2=np.random.randint(1, size=[20,2])

#DataFrame
df = pd.DataFrame({'a':a,'b':b})
```

```python
c=np.array([[20,21],[45,45]])
```

```python
plt.scatter(df.a, df.b)
plt.scatter(c[0],c[1], c=['g','r'], marker='x')
plt.show()
```

```python
for i in range(20):
    plt.scatter(df.a, df.b)
    plt.scatter(c[0],c[1], c=['g','r'], marker='x')
    plt.plot([c[0][0],a[i]],[c[1][0],b[i]],'--',c='g')
    clear_output(wait=True)
    plt.show()


clear_output(wait=True)
for i in range(20):
    plt.scatter(df.a, df.b)
    plt.scatter(c[0],c[1], c=['g','r'], marker='x')
    plt.plot([c[0][1],a[i]],[c[1][1],b[i]],'--',c='r')
    clear_output(wait=True)
    plt.show()
```

```python
plt.scatter(df.a, df.b)
for i in range(20):
    plt.scatter(c[0],c[1], c=['g','r'], marker='x')
    plt.plot([c[0][0],a[i]],[c[1][0],b[i]],'--',c='g')
    d1[i]=np.sqrt(((c[0][0]-a[i])**2) + ((c[1][0]-b[i])**2))
plt.show()

plt.scatter(df.a, df.b)
for i in range(20):
    plt.scatter(c[0],c[1], c=['r','y'], marker='x')
    plt.plot([c[0][1],a[i]],[c[1][1],b[i]],'--',c='r')
    d2[i]=np.sqrt(((c[0][1]-a[i])**2) + ((c[1][1]-b[i])**2))
plt.show()
```

```python
c0x=[];c0y=[];c1x=[];c1y=[]

plt.scatter(df.a, df.b)
plt.scatter(c[0],c[1], c=['g','r'], marker='x')
for i in range(20):
    if d1[i]<d2[i]:
        plt.plot([c[0][0],a[i]],[c[1][0],b[i]],'--',c='g')
        c0x.append(a[i])
        c0y.append(b[i])
    else:
        plt.plot([c[0][1],a[i]],[c[1][1],b[i]],'--',c='r')
        c1x.append(a[i])
        c1y.append(b[i])
plt.show()

print('Number of points in Green Cluster = {}'.format(len(c0x)))
print('Number of points in Red   Cluster = {}'.format(len(c1x)))
```

```python
print('New Centroids')
print(c)

plt.scatter(df.a, df.b)
o=copy.deepcopy(c)
plt.scatter(o[0],o[1], c=['g','r'], marker='x')

c[0][0]=sum(c0x)/len(c0x)
c[1][0]=sum(c0y)/len(c0y)
c[0][1]=sum(c1x)/len(c1x)
c[1][1]=sum(c1y)/len(c1y)


plt.scatter(c[0],c[1], c=['g','r'], marker='x')

plt.arrow(o[0][0],o[1][0],c[0][0]-o[0][0],c[1][0]-o[1][0],shape='full',lw=1, length_includes_head=True,zorder=0, head_length=2, head_width=1,color='g',linestyle = '--',)
plt.arrow(o[0][1],o[1][1],c[0][1]-o[0][1],c[1][1]-o[1][1],shape='full',lw=1, length_includes_head=True,zorder=0, head_length=2, head_width=1,color='r',linestyle = '--',)
plt.show()
```
